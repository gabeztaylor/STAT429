```python
import pandas as pd
from sodapy import Socrata
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
from datetime import datetime

from functools import reduce
```

```python
query = f"""SELECT date, primary_type, arrest  WHERE arrest=TRUE AND date > '2010-01-01T00:00:00.000' LIMIT 1000000"""
client = Socrata("data.cityofchicago.org", 'K7BDwBXpB1mz7gfKtyDY7QmL8')
results = client.get("ijzp-q8t2", query=query)
df_socrata = pd.DataFrame.from_records(results)
```

```python
df_socrata
```

```python
df_socrata['Date'] = list(map(lambda x: x[:10], df_socrata['date']))
df_socrata
```

```python
df_socrata['primary_type'].value_counts()
```

```python
df = df_socrata.groupby('Date').agg({'arrest' : 'sum'}).reset_index()
df
```

```python
to_date = lambda x: datetime.strptime(x, '%Y-%m-%d').date()
df['Date'] = list(map(to_date, df['Date']))
```

### Merge

```python
data_frames = [df[['Date', 'arrest']], temp[['Date', 'AWND', 'PRCP', 'TAVG']], stonk[['Date', 'Price']]]
```

```python
merge = lambda left, right: pd.merge(left, right, on = ['Date'], how = 'outer')
df_merged = reduce(merge, data_frames)
```

```python
df_merged = df_merged.sort_values('Date').iloc[:-6, :]
```

### Impute Price with previous price

```python
df_merged['Price'] = df_merged['Price'].fillna(method='ffill')
df_merged['Price'] = df_merged['Price'].fillna(method='bfill')
```

### Impute Weather Stats with rolling averages

```python
df_merged['TAVG'] = df_merged['TAVG'].fillna(df_merged['TAVG'].rolling(3, center=True, min_periods = 1).mean())
df_merged['AWND'] = df_merged['AWND'].fillna(df_merged['AWND'].rolling(3, center=True, min_periods = 1).mean())
df_merged['PRCP'] = df_merged['PRCP'].fillna(df_merged['PRCP'].rolling(3, center=True, min_periods = 1).mean())
```
